# In this course

We will derive, implement, evaluate, and theorize about learning algorithms - classical and modern

1. Introduce the notion of __statistical learning theory__ to guide our discussion and ground models in a common mathematical framework

2. Derive specific algorithms from different constraints, _inductive biases_, and intuitions about learnabaility

3. Implement and optimize learning algorithms in python without heavy reliance on libraries

4. Rigorously evaluate different algorithms empirically and theoretically when possible

5. Apply our knowledge to solve real world problems!

## Prereqs

You will be expected to know (or catch yourself up on) 
1. Statistical concepts: random variables, samples, mean, variance,
distributions, conditional variables and distributions, likelihood,
estimators and linear regression.
2. Linear algebra and multivariate calculus
3. Programming in python
